```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon = pd.read_csv('/content/Pokemon.csv', encoding='latin-1')
```

```python
pokemon.describe()
```

```python
fig, axs = plt.subplots(9, figsize = (12,12))
plt1 = sns.boxplot(pokemon['#'], ax = axs[0])
plt2 = sns.boxplot(pokemon['Total'], ax = axs[1])
plt3 = sns.boxplot(pokemon['HP'], ax = axs[2])
plt4 = sns.boxplot(pokemon['Attack'], ax = axs[3])
plt5 = sns.boxplot(pokemon['Defense'], ax = axs[4])
plt6 = sns.boxplot(pokemon['Sp. Atk'], ax = axs[5])
plt7 = sns.boxplot(pokemon['Sp. Def'], ax = axs[6])
plt8 = sns.boxplot(pokemon['Speed'], ax = axs[7])
plt9 = sns.boxplot(pokemon['Stage'], ax = axs[8])
plt.show()
```
